## Step 3: Merging DataFrames

Real-world data is often stored across multiple tables.
Using `pd.merge()`, we can combine tables based on a **common key column**.

```python
pd.merge(df_left, df_right, on='key', how='join_type')
```

### Join Types

inner join ‚Üí rows present in both tables
left join ‚Üí keep all rows from the left table
right join ‚Üí keep all rows from the right table
outer join ‚Üí keep all rows from both tables

### Practice Data: Bookstore System
- üì¶ **orders** : Order information
- üìö **books** : Book details
- üë§ **customers** : Customer information


````python
orders = pd.DataFrame({
    'order_id':    [1, 2, 3, 4, 5, 6, 7, 8],
    'book_id':     [101, 102, 101, 103, 104, 102, 105, 101],
    'customer_id': [201, 202, 203, 201, 204, 205, 202, None],
    'quantity':    [2, 1, 3, 1, 2, 1, 4, 1],
    'order_date':  ['2024-01-05','2024-01-07','2024-01-08','2024-01-10',
                    '2024-01-12','2024-01-15','2024-01-18','2024-01-20']
})

print("üì¶ orders")
orders
````

````python
books = pd.DataFrame({
    'book_id': [101, 102, 103, 104, 106],
    'title':   ['Harry-Potter', 'Lord of the Rings', 'Little Prince', 'Demian', 'Norwegian Wood'],
    'genre':   ['fantasy', 'fantasy', 'Literature', 'Literature', 'Novel'],
    'price':   [18000, 22000, 12000, 14000, 16000]
})

print("üìö books")
books
````

````python
customers = pd.DataFrame({
    'customer_id': [201, 202, 203, 204, 205, 206],
    'name':    ['Kim', 'Lee', 'Park', 'Choi', 'Chung', 'Han'],
    'age':     [28, 34, 22, 41, 30, 27],
    'city':    ['Seoul', 'Busan', 'Seoul', 'Daegu', 'Incheon', 'Gwangju'],
    'country': ['KR', 'KR', 'KR', 'KR', 'KR', 'KR']
})

print("üë§ customers")
customers
````

### 3-1. Inner Join ‚Äî Only Matching Data

````python
# Inner join: orders + books
merged_inner = pd.merge(orders, books, on='book_id', how='inner')

print(f"rows of orders       : {len(orders)}")
print(f"rows after inner join : {len(merged_inner)}")
print("book_id=105 is excluded \n")
merged_inner
````

### 3-2. Left Join ‚Äî Based on Left Table

````python
merged_left = pd.merge(orders, books, on='book_id', how='left')

print(f"rows after left join: {len(merged_left)}")
print("‚Üí book_id=105 is included, but title/genre/price are NaN!\n")
merged_left
````

### 3-3. Second Merge ‚Äî Add Customer Information

````python
# Translated comment
final_df = pd.merge(merged_inner, customers, on='customer_id', how='left')

print(f"final shape: {final_df.shape}")
print()
final_df
````

## Step 4: Checking Missing Values (Null)

Data may contain missing values (`NaN`).
Before analysis, we should identify and decide how to handle them.

| Code | Description |
|------|-------------|
| `df.isnull().sum()` | Number of missing values |
| `df.isnull().mean() * 100` | Percentage of missing values |
| `df[df['col'].isnull()]` | Filter rows with null |
| `df[df['col'].notnull()]` | Filter rows without null |


````python

print(" N of Null:")
titanic.isnull().sum()
````

````python
null_ratio = (titanic.isnull().sum() / len(titanic) * 100).round(1)
print("Null ratio (%):")
null_ratio[null_ratio > 0]
````

````python

age_null_df = titanic[titanic['Age'].isnull()]

print(f"N of no age information: {len(age_null_df)}Î™Ö")
age_null_df[['Survived','Pclass','Sex','Age','Fare']].head(10)
````

````python

print("Í≥†Í∞ù IDÍ∞Ä ÏóÜÎäî Ï£ºÎ¨∏:")
orders[orders['customer_id'].isnull()]
````

## Step 6: Conditional Filtering

Selecting rows based on conditions is one of the most common tasks in data analysis.

```python
df[df['column'] == 'value']
df[df['column'] != 'value']
df[df['column'] > number]

df[(cond1) & (cond2)]
df[(cond1) | (cond2)]
```

> ‚ö†Ô∏è Wrap each condition in parentheses when combining multiple conditions.


````python

first_class = titanic[titanic['Pclass'] == 1]

print(f"N of first class: {len(first_class)}Î™Ö")
first_class[['Survived','Pclass','Sex','Age','Fare']].head(8)
````

````python

survived_1st_female = titanic[
    (titanic['Pclass'] == 1) &
    (titanic['Sex'] == 'female') &
    (titanic['Survived'] == 1)
]

print(f"N of female in the first_class : {len(survived_1st_female)} ")
survived_1st_female[['Survived','Pclass','Sex','Age','Fare']].head()
````

````python

elderly = titanic[titanic['Age'] >= 60]

print(f"N of passengers over 60: {len(elderly)}Î™Ö")

survival_rate = elderly['Survived'].mean() * 100
print(f"Survival rate of passengers over 60: {survival_rate:.1f}%")
````

````python

sq_passengers = titanic[
    (titanic['Embarked'] == 'S') |
    (titanic['Embarked'] == 'Q')
]

print(f"N of embarking from Southampton or Queenstown: {len(sq_passengers)}Î™Ö")
````

````python

fantasy_orders = final_df[final_df['genre'] == 'fantasy']

print(f"N of Fantasy orders: {len(fantasy_orders)}")
fantasy_orders
````

## Step 8: Grouped Aggregation (GroupBy)

`groupby()` groups data based on a column and applies aggregation functions.

```python
df.groupby('group')['target'].mean()
```

Common aggregation functions:
- `.mean()`
- `.median()`
- `.sum()`
- `.count()`
- `.max()` / `.min()`


````python
print("average age group by pclass:")
print(titanic.groupby('Pclass')['Age'].mean().round(1))
print()
print("median age group by pclass:")
print(titanic.groupby('Pclass')['Age'].median().round(1))
````

````python

survival_by_sex = titanic.groupby('Sex')['Survived'].mean() * 100
print("survival rate group by sex (%):")
survival_by_sex.round(1)
````

````python
# Translated comment
print("average book price group by genre:")
final_df.groupby('genre')['price'].mean()
````

````python
# Translated comment
print("sum of order quantity group by city:")
final_df.groupby('city')['quantity'].sum().sort_values(ascending=False)
````

````python
# Translated comment
print("Survival rate of Pclass √ó Sex (%):")
(titanic.groupby(['Pclass','Sex'])['Survived'].mean() * 100).round(1)
````
