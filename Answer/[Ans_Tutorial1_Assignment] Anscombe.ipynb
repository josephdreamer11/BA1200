```python
fig, ax = plt.subplots(2,2, sharex = 'col', sharey = 'row')

def calc(data):
    mean = data.mean()
    variance = data.var()
    std = data.std()
    return mean, variance, std


def poly(x,y, subplot) :
    mean, variance, std = calc(y)
    slope, intercept = np.polyfit(x, y, 1)
    subplot.scatter(x, y)
    subplot.plot(x, slope*x + intercept, color = 'red')
    subplot.legend([f'Mean: {mean:.2f}\nvariance: {variance:.2f} \n std : {std:.2f}'])

poly(df['x123'],df['y1'],ax[0][0])
poly(df['x123'],df['y2'],ax[0][1])
poly(df['x123'],df['y3'],ax[1][0])
poly(df['x4'],df['y4'],ax[1][1])

```
